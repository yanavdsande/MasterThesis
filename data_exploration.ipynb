```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('/content/data_chimp_videos.xlsx')
```

```python
df.head()
```

```python
df['Duration_string'] = (df['Duration_in_mm_ss'].astype(str))
df['Duration_correct_format'] = [ "00:"+i[11:16] for i in df['Duration_string'] ]

print(df[114:116])
```

```python
df['timedeltas'] = pd.to_timedelta(df['Duration_correct_format'])
df.head()
```

```python
df_good_quality = df.loc[df['Good_quality'].isin([2,3])]
df_good_quality.head()
```

```python
print(f"This is the amount of videos we have: {len(df['Video_names'])}")
```

```python
print(f"This is the total duration of videos we have: {df['timedeltas'].sum()}")
```

```python
print(f"This is the amount of high quality videos we have: {len(df_good_quality)}")
```

```python
print(f"The total duration of high quality video data is: { df_good_quality['timedeltas'].sum()}")
```

```python
print(f"In the dataset we have: {len(df_good_quality['Contrast']=='Low')} videos of low contrast, {len(df_good_quality['Contrast']=='High')} of high contrast and {len(df_good_quality['Contrast'] =='Medium')} videos of medium contrast")
```
